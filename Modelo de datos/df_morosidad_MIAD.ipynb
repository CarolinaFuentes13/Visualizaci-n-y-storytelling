```python
# Lista de columnas finales
columnas_finales = ['Llave2', 'Nombre_linea', 'Fecha_aprobacion',
       'Antiguedad_meses', 'Nombre_fondo', 'Valor_financiacion', 'Cuotas',
       'Tipo_interes', 'Vr_neto_matricula', 'Y_categorica', 'IDBANNER',
       'FECHA_NACIMIENTO', 'ESTADO_CIVIL', 'DIRECCION', 'CIUDAD', 'CODCIUDAD',
       'GENERO', 'FACULTAD', 'PROGRAMA', 'NIVEL', 'ESTADO',
       'TIPOESTUDIANTE', 'SEDE', 'SELLO', 'CARGA', 'GRUPO_ETNICO',
       'TIPO_DISCAPACIDAD', 'NACIONALIDAD',
       'Mora', 'Valor_cuota_inicial', 'Valor_primera_cuota',
       'fecha de pago', 'Dirección_Scoring', 'Ciudad_Scoring',
       'Departamento_Scoring', 'Validación valor financiado',
       'Detalle estado final', 'ID Estudiante', 'Tipo_estudiante',
       'Operación', 'Cate', 'Subcate', 'Cohorte', 'Mes', 'Cliente',
       'Media Score', 'AñoB', 'VALOR MAXIMO', 'VALOR MEDIO', 'VALOR BAJO',
       'Plataforma']


df_final= df_inicial[columnas_finales].copy()

df_final
```

```python
df_final.info()
```

```python
#eliminamos filas con IDBANNER o ID Estudiante nulos
df_final = df_final[df_final['IDBANNER'].notnull()]
df_final = df_final[df_final['ID Estudiante'].notnull()]
```

```python
#Diferencias en las dos columnas de direcciones
df_final[['DIRECCION', 'Dirección_Scoring']].isnull().sum()
```

```python
#Imputamos entre ellas
df_final['DIRECCION'] = df_final['DIRECCION'].fillna(df_final['Dirección_Scoring'])
df_final['DIRECCION'].isnull().sum()
```

```python
#Llenamos los valores faltantes restantes con 'NO REGISTRADA'
df_final['DIRECCION']= df_final['DIRECCION'].fillna('NO REGISTRADA')
df_final['DIRECCION'].isnull().sum()
```

```python
#Lo eliminamos del dataframe final
df_final.drop(columns=['Dirección_Scoring'], inplace=True)
df_final.columns
```

```python
#Diferencias en las dos columnas de ciudades
df_final[['CIUDAD', 'Ciudad_Scoring']].isnull().sum()
```

```python
#Imputamos entre ellas
df_final['CIUDAD'] = df_final['CIUDAD'].fillna(df_final['Ciudad_Scoring'])
df_final['CIUDAD'].isnull().sum()
```

```python
mapa_codciu = (
    df_final[['CODCIUDAD', 'CIUDAD']]
    .dropna()
    .drop_duplicates(subset=['CODCIUDAD'])
    .set_index('CODCIUDAD')['CIUDAD']
)
```

```python
#Imputamos los valores faltantes restantes usando el CODCIUDAD
df_final.loc[df_final['CIUDAD'].isnull(), 'CIUDAD'] = \
    df_final.loc[df_final['CIUDAD'].isnull(), 'CODCIUDAD'].map(mapa_codciu)
```

```python
df_final['CIUDAD'].isnull().sum()
```

```python
df_final['CODCIUDAD'].isnull().sum()
```

```python
ciudades_sin_codigo = df_final.loc[df_final['CODCIUDAD'].isnull(), 'CIUDAD'].unique()
map_ciudad_cod = (
    df_final[['CIUDAD', 'CODCIUDAD']]
    .dropna()
    .drop_duplicates()
    .set_index('CIUDAD')['CODCIUDAD']
    .to_dict()
)
```

```python
#imputamos los valores faltantes de CODCIUDAD usando CIUDAD
df_final['CODCIUDAD'] = df_final['CODCIUDAD'].fillna(
    df_final['CIUDAD'].map(map_ciudad_cod))
```

```python
df_final['CODCIUDAD'].isnull().sum()
```

```python
df_nulos_codigo= df_final[
    df_final["CODCIUDAD"].isnull()]

df_nulos_codigo = df_nulos_codigo[ 
    ["NACIONALIDAD", "CODCIUDAD", 
     "CIUDAD"]]

df_nulos_codigo
```

```python
#Cargamos el archivo DIVIPOLA para hacer la imputación 
Divipola = pd.read_excel('DIVIPOLA_Municipios.xlsx')
Divipola
```

```python
# Normalizar nombres
df_final['CIUDAD_norm'] = df_final['CIUDAD'].str.upper().str.strip()
Divipola['Nombre_mun_norm'] = Divipola['Nombre_mun'].str.upper().str.strip()
```

```python
Divipola.columns    
```

```python
map_divipola = (
    Divipola[['Nombre_mun_norm', 'Código _mun']]
    .dropna()
    .drop_duplicates()
    .set_index('Nombre_mun_norm')['Código _mun']
    .to_dict())
```

```python
#imputamos los valores faltantes de CODCIUDAD usando CIUDAD del DIVIPOLA
df_final['CODCIUDAD'] = df_final['CODCIUDAD'].fillna(
    df_final['CIUDAD_norm'].map(map_divipola))
```

```python
df_final['CODCIUDAD'].isnull().sum()
```

```python
df_final.isnull().sum()[df_final.isnull().sum() > 0]
```

```python
#agregamos el nombre del departamento usando el DIVIPOLA
df_final = df_final.merge(
    Divipola[["Código _mun", "Nombre_dep"]],
    left_on="CODCIUDAD",
    right_on="Código _mun",
    how="left")
```

```python
#Imputamos los valores faltantes de Departamento_Scoring usando Nombre_dep
df_final['Departamento']=df_final["Departamento_Scoring"] = (
    df_final["Departamento_Scoring"]
    .fillna(df_final["Nombre_dep"]))
```

```python
df_final["Departamento"].isnull().sum()
```

```python
df_final = df_final.drop(columns=["Nombre_dep", "Ciudad_Scoring"])
df_final.columns
```

```python
#agregamos Longitud y Latitud usando el DIVIPOLA
df_final = df_final.merge(
    Divipola[['Código _mun', 'Longitud', 'Latitud']],
    how='left',
    left_on='Código _mun',
    right_on='Código _mun')
df_final[['Longitud', 'Latitud']].isnull().sum()
```

```python
#Identificamos las ciudades que aún no tienen Longitud y Latitud
ciudades_sin_geo = df_final.loc[df_final['Longitud'].isnull(), 'CIUDAD_norm'].unique()
ciudades_sin_geo
```

```python
#Revisamos la base de DIVIPOLA manual para ver si hay errores y corregímos
correcciones = {
    'BELEN DE BAJIRA': 'NUEVO BELÉN DE BAJIRÁ'}

df_final['CIUDAD_norm'] = df_final['CIUDAD_norm'].replace(correcciones)
```

```python
df_final['Código _mun'].isnull().sum()
```

```python
#Identificamos las filas con Código _mun faltante
faltantes = df_final[df_final['Código _mun'].isnull()][['CIUDAD_norm']]
faltantes_merge = faltantes.merge(
    Divipola[['Nombre_mun', 'Código _mun', 'Longitud', 'Latitud']],
    how='left',
    left_on='CIUDAD_norm',
    right_on='Nombre_mun')
```

```python
#Imputamos los valores faltantes de Código _mun, Longitud y Latitud
df_final.loc[df_final['Código _mun'].isnull(), 'Código _mun'] = faltantes_merge['Código _mun'].values
df_final.loc[df_final['Longitud'].isnull(), 'Longitud'] = faltantes_merge['Longitud'].values
df_final.loc[df_final['Latitud'].isnull(), 'Latitud'] = faltantes_merge['Latitud'].values
```

```python
df_final[['Código _mun', 'Longitud', 'Latitud']].isnull().sum()
```

```python
#limpiamos el dataframe final del modelo
df_final = df_final.drop(columns=['IDBANNER','Departamento_Scoring','CIUDAD', 'CODCIUDAD'])
df_final = df_final.reset_index(drop=True)
df_final.columns
```

```python
#final revisión de nulos 
df_final.isnull().sum()[df_final.isnull().sum() > 0]
```

```python
#Estandarizamos los nombres de las columnas
import unicodedata

def estandarizar_columna(col):
    col = unicodedata.normalize('NFKD', col).encode('ascii', 'ignore').decode('utf-8')
    col = col.lower()
    col = col.replace(" ", "_")
    col = col.replace("-", "_").replace("/", "_")
    col = "_".join(part for part in col.split("_") if part)
    return col
df_final.columns = [estandarizar_columna(col) for col in df_final.columns]
```
